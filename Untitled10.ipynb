```python
df.drop_duplicates(inplace=True)
```

```python
avg=df["flow_duration"].mean()
avg
```

```python
df
```

```python
df['flow_duration'].fillna(avg,inplace=True)
df
```

```python
df['total_forward_packets'].fillna("Not Mentioned",inplace=True)
df
```

```python
df['total_backward_packets'].fillna(0,inplace=True)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['total_forward_packets'], df['flow_duration'], color='Green')
plt.title("Forward packets VS Flow duration")
plt.xlabel("total_forward_packets")
plt.ylabel('flow_duration')
plt.show()
```

```python
plt.bar(df['total_forward_packets'], df['flow_duration'], color='Green')
plt.title("Forward packets length VS Flow duration")
plt.xlabel("Forward packets length")
plt.ylabel("Flow durations")
plt.show()
```

```python
plt.plot(df['total_forward_packets'], df['total_backward_packets'], color='Green')
plt.title("Forward VS Backward packets")
plt.xlabel("total_forward_packets")
plt.ylabel("total_backward_packets")
plt.show()
```

```python
plt.pie(df['flow_duration'],labels=df['total_forward_packets'])
plt.xticks(rotation=25)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.xticks(rotation=25)
plt.pie(df['flow_duration'],labels=df['total_forward_packets'])
plt.suptitle("Flow duration VS Total forward packets")
plt.xlabel("Flow duration")
plt.ylabel("Total forward packets")

plt.subplot(2,2,2)
plt.xticks(rotation=25)
plt.scatter(df['total_forward_packets'],df['total_backward_packets'],color='green')
plt.suptitle("Forward VS Backward packets")
plt.xlabel("total forward packets")
plt.ylabel("total backard packets")



plt.subplot(2,2,3)
plt.xticks(rotation=25)
plt.plot(df['total_forward_packets'], df['total_backward_packets'], color='Green')
plt.suptitle("Forward VS Backward packets")
plt.xlabel("total forward packets")
plt.ylabel("total backward packets")



plt.subplot(2,2,4)
plt.xticks(rotation=25)
plt.bar(df['total_forward_packets'], df['flow_duration'], color='Green')
plt.suptitle("Total forward packets VS Flow duration")
plt.xlabel("Forward packets")
plt.ylabel("Flow packets")

plt.tight_layout()


plt.savefig("DashBoard.png")
files.download("DashBoard.png")
```
